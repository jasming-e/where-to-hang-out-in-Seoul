# 2. 관광 + 날씨 데이터 합치기

```python
df_t = df_t.rename(columns = {'구' : '시군구명'})
```

```python
gu = gu.rename(columns = {'한국어' : '시군구명'})
gu
```

```python
df_tw_merged = pd.merge(df_tw, gu, how = 'right', on = '시군구명')
df_tw_merged
```

```python
df_t1 = df_t[df_t['언어'] == '한국어']
df_t1_merged = pd.merge(df_t1, df_tw_merged , how = 'inner', on = '시군구명')
df_t1_merged
```

```python
df_t2 = df_t[df_t['언어'] == 'ENGLISH']
df_t2 = df_t2.rename(columns = {'시군구명' : '영어'})
```

```python
df_t2_merged = pd.merge(df_t2, df_tw_merged, how = 'inner', on = '영어')
df_t2_merged
```

```python
df_t3 = df_t[df_t['언어'] == '汉语']
df_t3 = df_t3.rename(columns = {'시군구명' : '중국어'})
df_t3_merged = pd.merge(df_t3, df_tw_merged, how = 'inner', on = '중국어')
df_t3_merged
```

```python
df_t4 = df_t[df_t['언어'] == '日本語']
df_t4 = df_t4.rename(columns = {'시군구명' : '일본어'})
df_t4_merged = pd.merge(df_t4, df_tw_merged, how = 'inner', on = '일본어')
df_t4_merged
```

```python
df_total = pd.concat([df_t1_merged,df_t2_merged,df_t3_merged,df_t4_merged])
df_total
```

```python
df_total.to_excel('/content/drive/MyDrive/빅데이터통계분석/data/seoul_total.xlsx', index = 'None')

from google.colab import files
files.download('/content/drive/MyDrive/빅데이터통계분석/data/seoul_total.xlsx')
```
