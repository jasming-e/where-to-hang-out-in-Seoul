# 2. GET OPENAPI

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
html = urlopen(myurl)
print(html.read().decode('UTF-8'))
```

```python
html = urlopen(myurl)
bs = BeautifulSoup(html.read(), 'lxml-xml')
```

```python
itemlist = bs.find_all('item')

df = pd.DataFrame()

for t in itemlist :
    tm = t.find('tm').get_text()
    totalcityname = t.find('totalCityName').get_text()
    doname = t.find('doName').get_text()
    cityname = t.find('cityName').get_text()
    cityid = t.find('cityAreaId').get_text()
    kmaTci = t.find('kmaTci').get_text()
    tciGrade = t.find('TCI_GRADE').get_text()

    temp = pd.DataFrame(([[tm, totalcityname , doname, cityname, cityid, kmaTci, tciGrade]]),
                        columns= ['일자', '주소', '시도명' ,'시군구명', 'id', '관광지수', '관광등급'])
    df = pd.concat([df,temp])
```

```python
df
```

```python
df = df[df['시도명'] == '서울']
df
```

```python
df = df.sort_values(by = ['시군구명', '일자'])
df = df.reset_index(drop= True)
df
```
